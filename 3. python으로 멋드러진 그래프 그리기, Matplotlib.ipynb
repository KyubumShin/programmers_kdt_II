## I. Matplotlib 시작하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## II. Matplotlib Case Study

```python
#명시안하면 x=index y = [1,2,3,4,5]
```

```python
plt.figure(figsize=(6,6))
plt.plot([0,1,2,3,4,5])
plt.show()
```

## 2차함수 그래프 with plot()

```python
plt.plot([0,1,2,3,4])
plt.show()
```

```python
x = np.arange(-10,10,0.01)
y = x**2
#x,y 축 설명 붙이기
plt.xlabel("x value")
plt.ylabel("f(x) value")

# x,y축 범위 설정하기
plt.axis([-5,5,0,25])

#x,y축 눈금 설정하기
plt.xticks([i for i in range(-5,6,1)])
plt.yticks([i for i in range(0,26,3)])

#제목 설정
plt.title("y = x^2 graph")

#범례 이름설정
plt.plot(x, y, label="trend")

#범례설정
plt.legend()
plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프(Plot)
* .plot()

```python
x = np.arange(20)
y = np.random.randint(0,20,20)

plt.axis([0,20,0,20])
plt.yticks([i for i in range(0,21,5)])

plt.plot(x,y)
plt.show()
```

### 산점도 (Scatter Plot)
* .scatter()

```python
plt.scatter(x,y)
plt.show()
```

### 박스 그림 (Box Plot)
*.boxplot()

```python
plt.boxplot((x,y))
plt.title("Box plot of x,y")
plt.show()
```

### 막대 그래프 (Bar Plot)
* 범주형 데이터의 값과 그값의 크기를 직사각형으로 나타낸 그림  
도수를 보기 용이

```python
plt.bar(x,y)
plt.xticks(np.arange(0,20,2))

plt.show()
```

```python
plt.hist(y, bins = np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

### 원형 그래프(Pie Chart)
비율을 보기에 용이
* .pie()

```python
z=  [100,200,300,200]
plt.pie(z, labels=["1","2","3","4"])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

```python
import seaborn as sns
```

### 커널밀도그림(Kernel Density Plot)
* 히스토그램과 같은 연속적인 분포를 곡선화 해서 그린 그림
* sns.kdeplot()

```python
x = np.arange(0,22,2)
y = np.random.randint(0, 20, 20)
plt.hist(y, bins = x)
plt.show()
```

```python
sns.kdeplot(y, shade = True)
plt.show()
```

### 카운트 그림(Count Plot)
* 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는것과 동일한 효과
* sns.countplot()

```python
vote_df = pd.DataFrame({"name":["Andy","Bob", "Cat"], "vote":[True,False,True]})
vote_df
```

```python
vote_count = vote_df.groupby("vote").count()
vote_count
```

```python
plt.bar(x= [False, True], height=vote_count["name"])
plt.show()
```

```python
sns.countplot(x = vote_df["vote"])

plt.show()
```

### 캣그림(Cat Plot)
* 숫자형 변수와 하나이상의 범주형 변수의 관계를 보여주는 함수
* sns.catplot()

```python
covid = pd.read_csv("./country_wise_latest.csv")
covid.head(5)
```

```python
s = sns.catplot(x="WHO Region", y="Confirmed", data=covid)
s.fig.set_size_inches(10,6)
plt.show()
```

### 스트립 그림 (Strip plot)
* scatter plot 과 유사하게 데이터의 수치를 표현하는 그래프
* sns.stripplot()

```python
sns.stripplot(x='WHO Region',y='Recovered', data=covid)
```

```python
#cf) swarmplot
sns.swarmplot(x='WHO Region',y='Recovered', data=covid)
```

### Heatmap
* 데이터의 행렬을 색상으로 표현해주는 그래프


```python
sns.heatmap(covid.corr())
```
